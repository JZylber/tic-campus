# Ciclos en Python

Para aprender ciclos en python, vamos a hacer un programa que cuente la cantidad de una determinada letra en una palabra

## While

```python
palabra = "inconvenience"
letra = 'i'
indice = 0
cantidad = 0
while indice < len(palabra):
    if palabra[indice] == letra:
        cantidad += 1
    indice += 1
print(f"La letra '{letra}' aparece {cantidad} veces en la palabra '{palabra}'")
```

## For (por índices)

```python
palabra = "inconvenience"
letra = 'i'
cantidad = 0
for indice in range(0, len(palabra)):
    if palabra[indice] == letra:
        cantidad += 1
print(f"La letra '{letra}' aparece {cantidad} veces en la palabra '{palabra}'")
```

## For (for each o de iteradores)

```python
palabra = "inconvenience"
letra = 'i'
cantidad = 0
for letra_palabra in palabra:
    if letra_palabra == letra:
        cantidad += 1
print(f"La letra '{letra}' aparece {cantidad} veces en la palabra '{palabra}'")
```
