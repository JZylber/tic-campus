## Buscando a Wally
Dada una lista de nombres, ¿Podemos encontrar a Wally? ¿En que posición está?

```python
nombres = ["Alicia", "Bernardo", "Carla", "Wally", "David", "Elena"]
encontrado = False
for pos,nombre in enumerate(nombres):
    if nombre == "Wally":
        print(f"Encontrado en la posición {pos + 1}")
        encontrado = True
        break
if not encontrado:
    print("No se encontró a Wally")
```
