# Introducción a las listas en Python

Las **listas** son una de las estructuras de datos más usadas en Python.
Sirven para guardar varios valores dentro de una misma variable y mantenerlos en un **orden**.

Por ejemplo, en lugar de tener variables separadas como:

```python
nombre_1 = "Ana"
nombre_2 = "Luis"
nombre_3 = "Carla"
```

podemos escribir:

```python
nombres = ["Ana", "Luis", "Carla"]
```

Eso hace que el código sea más cómodo de leer, más fácil de recorrer y mucho más práctico para trabajar con colecciones de datos.

En este apunte vamos a ver:

- cómo crear listas,
- cómo acceder a sus elementos,
- cómo recorrerlas,
- cómo usar `enumerate`,
- cómo trabajar con **listas de listas**,
- y cómo buscar elementos dentro de una lista.


## Crear listas

En Python, una lista se crea usando corchetes `[]` y separando sus elementos con comas.

Las listas pueden contener números, textos y, en general, cualquier tipo de dato.
En este apunte vamos a usar **anotaciones de tipo** para que el código sea más claro.

Por ejemplo, una lista de enteros se puede escribir así:


````python
numeros: list[int] = [1, 2, 3, 4]
print(numeros)
````

También podemos tener una lista de strings:


````python
nombres: list[str] = ["Ana", "Luis", "Carla"]
print(nombres)
````

## Acceder a elementos por índice

Cada elemento de una lista tiene una posición llamada **índice**.

Lo importante es recordar que en Python los índices empiezan en **0**.

Entonces, en la lista:

```python
letras = ["a", "b", "c"]
```

- `"a"` está en la posición `0`
- `"b"` está en la posición `1`
- `"c"` está en la posición `2`

Podemos acceder a un elemento escribiendo el nombre de la lista seguido del índice entre corchetes.


````python
letras: list[str] = ["a", "b", "c"]

print(letras[0])
print(letras[2])
````

## Recorrer una lista con `for`

Una de las operaciones más comunes con listas es **recorrer todos sus elementos**.

Para eso usamos un `for`. En cada vuelta del loop, la variable toma el valor de un elemento de la lista.

Esto es útil cuando queremos imprimir, procesar o revisar todos los elementos uno por uno.


````python
valores: list[int] = [10, 20, 30]

for valor in valores:
    print(valor)
````

## Recorrer una lista con índice usando `enumerate`

A veces no solo queremos el valor, sino también su posición.

Para eso usamos `enumerate`, que nos devuelve dos cosas en cada iteración:

1. el índice,
2. el elemento.

Esto es muy útil cuando queremos informar en qué posición apareció algo.


````python
nombres: list[str] = ["Ana", "Luis", "Carla"]

for posicion, nombre in enumerate(nombres):
    print(posicion, nombre)
````

## Listas de listas

Una lista también puede contener **otras listas**.

Eso se usa mucho para representar estructuras más complejas, por ejemplo una tabla o una matriz.

En este caso, cada fila es una lista, y la lista principal contiene a todas las filas.

Podemos anotar el tipo de una matriz de enteros como `list[list[int]]`.


````python
matriz: list[list[int]] = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(matriz)
````

Si queremos acceder a un elemento de una lista de listas, usamos dos índices:

- el primero indica la fila,
- el segundo indica la columna.

Por ejemplo, `matriz[0][1]` significa:

- fila `0` → `[1, 2, 3]`
- columna `1` dentro de esa fila → `2`


````python
matriz: list[list[int]] = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(matriz[0])
print(matriz[0][1])
print(matriz[2][2])
````

También podemos recorrer una lista de listas.
En ese caso, primero recorremos las filas y después, si queremos, los elementos de cada fila.


````python
matriz: list[list[int]] = [
    [1, 2, 3],
    [4, 5, 6]
]

for fila in matriz:
    print("Fila:", fila)

    for elemento in fila:
        print("Columna:", elemento)
````

## Búsqueda lineal

La forma más simple de buscar un elemento en una lista es revisar los elementos uno por uno desde el principio.

A este método se lo suele llamar **búsqueda lineal**.

La función de abajo recibe:

- una lista de enteros,
- un número que queremos buscar,

y devuelve la posición donde se encuentra.
Si no aparece, devuelve `-1`.


````python
def busqueda_lineal(lista: list[int], objetivo: int) -> int:
    for i, valor in enumerate(lista):
        if valor == objetivo:
            return i

    return -1


datos: list[int] = [3, 5, 7, 9]
print(busqueda_lineal(datos, 5))
print(busqueda_lineal(datos, 8))
````

## Búsqueda binaria (versión iterativa)

La búsqueda binaria es una forma más eficiente de buscar, pero tiene una condición importante:

> la lista debe estar **ordenada**.

La idea es comparar el valor buscado con el elemento del medio:

- si lo encontramos, terminamos,
- si el objetivo es más grande, seguimos buscando a la derecha,
- si el objetivo es más chico, seguimos buscando a la izquierda.

En esta versión usamos un `while`, no recursión.


````python
def busqueda_binaria(lista: list[int], objetivo: int) -> int:
    izquierda: int = 0
    derecha: int = len(lista) - 1

    while izquierda <= derecha:
        medio: int = (izquierda + derecha) // 2

        if lista[medio] == objetivo:
            return medio
        elif lista[medio] < objetivo:
            izquierda = medio + 1
        else:
            derecha = medio - 1

    return -1


ordenados: list[int] = [1, 2, 3, 4, 5, 6, 7, 8]
print(busqueda_binaria(ordenados, 6))
print(busqueda_binaria(ordenados, 10))
````

## Ejemplo: “Buscando a Wally”

Para cerrar, hagamos una función simple con listas de strings.

La idea es recorrer una lista y devolver un mensaje indicando en qué posición aparece `"Wally"`.

Este ejemplo sirve para practicar:

- recorrido de listas,
- uso de `enumerate`,
- comparación de strings,
- y retorno de resultados desde una función.


````python
def buscar_wally(lista: list[str]) -> str:
    for posicion, elemento in enumerate(lista):
        if elemento == "Wally":
            return f"Se encontro a wally en la posicion {posicion}"

    return "No se encontro a wally"


personajes: list[str] = ["Ana", "Mario", "Wally", "Luigi"]
print(buscar_wally(personajes))
print(buscar_wally(["Ana", "Luis", "Carla"]))
````

## Ejemplo: “Buscando a Wally” en una lista de listas

Aquí tenemos una versión de la función `buscar_wally` que funciona con una **lista de listas** (una matriz). En este caso, la función devolverá la fila y la columna donde se encuentra "Wally".

````python
def buscar_wally_en_matriz(matriz: list[list[str]]) -> str:
    for fila_idx, fila in enumerate(matriz):
        for col_idx, elemento in enumerate(fila):
            if elemento == "Wally":
                return f"Se encontro a Wally en la fila {fila_idx} y columna {col_idx}"
    return "No se encontro a wally"


ciudad: list[list[str]] = [
    ["Ana", "Pedro", "Maria"],
    ["Luis", "Wally", "Carla"],
    ["Juan", "Marta", "Jose"]
]

ciudad_sin_wally: list[list[str]] = [
    ["Ana", "Pedro"],
    ["Luis", "Carla"]
]

print(buscar_wally_en_matriz(ciudad))
print(buscar_wally_en_matriz(ciudad_sin_wally))
````
